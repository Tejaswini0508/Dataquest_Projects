```python
#Read the data
post_data=pd.read_csv("2019_questions.csv", parse_dates=["CreationDate"])
```

From the above we can see that the FavoriteCount column has null values.Also Tags column has some characters which we might need to clean.

Favorite columm will have Null Values because it might happen that the particular question is not in anyone's favourite list.We can replace the null values by zero.

```python
(post_data["Tags"].apply(lambda val:type(val))).unique()
```

# Cleaning the data

```python
# Replace Null Values with 0 in FavoriteCount column
post_data["FavoriteCount"].fillna(0,inplace=True)

post_data["FavoriteCount"]=post_data["FavoriteCount"].astype(int)
```

```python
#Change the data type of Tags Column to string
post_data["Tags"]=post_data["Tags"].astype(str)
```

```python
post_data.info()
```

```python
# Clean the tags column to get a comma seperated list of tags
post_data["Tags"]=post_data["Tags"].str.replace("><",",")
post_data["Tags"]=post_data["Tags"].str.replace(">","")
post_data["Tags"]=post_data["Tags"].str.replace("<","")
post_data["Tags"].head()
```

# Identifying Most Popular Tags

We will identify the most used tags and most viewed tags

## Most Used Tags

```python
post_data["Tags"]=post_data["Tags"].str.split(",")
```

```python
tags_used={}

for i in post_data["Tags"]:
    for j in i:
        if j in tags_used:
            tags_used[j]+=1
        else:
             tags_used[j]=1  
# Convert tags_used to a dataframe
tags_used = pd.DataFrame.from_dict(tags_used, orient="index")
tags_used.rename(columns={0: "Count"}, inplace=True)
```

```python
# Get the top 20 most used tags 

tags_used_top_20=tags_used.sort_values(by="Count").tail(20)

tags_used_top_20
```

```python
# Plot the top 20 most used tags
tags_used_top_20.plot(kind="barh",title="Top 20 most used tags",figsize=(16,8))

plt.show()
```

## Most Viewed Tags

```python
tags_viewed = dict()

for index, row in post_data.iterrows():
    for tag in row['Tags']:
        if tag in tags_viewed:
            tags_viewed[tag] += row['ViewCount']
        else:
            tags_viewed[tag] = row['ViewCount']
            
tags_viewed = pd.DataFrame.from_dict(tags_viewed, orient="index")
tags_viewed.rename(columns={0: "ViewCount"}, inplace=True)

tags_viewed_top_20 = tags_viewed.sort_values(by="ViewCount").tail(20)

tags_viewed_top_20.plot(kind="barh", figsize=(16,8))
```

# Relations between tags

One solution of finding how tags are related to each other is by counting the number of times they appear together.

Create a dataframe like below:

| Tags 	| Tag1 	| Tag2 	| Tag3 	|
|------	|------	|------	|------	|
| Tag1 	| NaN  	| NaN  	| NaN  	|
| Tag2 	| NaN  	| NaN  	| NaN  	|
| Tag3 	| NaN  	| NaN  	| NaN  	|

Further we caan fill in the values with the count of the respective tags appearing together.


```python
# Get the distinct tag names
tags_list=tags_used.index

# Create the data frame as above
tags_relation=pd.DataFrame(index=tags_list,columns=tags_list)

#Fill the NA values with 0

tags_relation.fillna(0,inplace=True)


tags_relation.head()
```

```python
# Populate the above dataframe with their respective count

for i in post_data["Tags"]:
    tags_relation.loc[i,i]+=1    
```

```python
tags_relation.head()
```

The above dataframe is too big to interpret.Hence we will only focus on the top 20 most_used tags.

```python
top_20_tag_relations=tags_relation.loc[tags_used_top_20.index,tags_used_top_20.index]
```

```python
top_20_tag_relations
```

It is very difficult to interpret and read the above dataframe. We will try to create the above data in form if a heatmap.

```python
#Plot the heatmap

plt.figure(figsize=(12,8))
sns.heatmap(top_20_tag_relations, cmap="Greens", annot=False)
```

We can see from the above heatmap that the data got skewed because of the diagonal values which tell the number of times the individual tag was used.Since we are only interested in pairs of tag we will make this values to be nulls.

```python
for i in range(top_20_tag_relations.shape[0]):
    top_20_tag_relations.iloc[i,i]=pd.np.NaN
    
```

```python
#Plot the heatmap

plt.figure(figsize=(12,8))
sns.heatmap(top_20_tag_relations, cmap="Greens", annot=False)
```

From the above we can see that the most used tags have strong relationships.But that also could be due to the fact that they are all so extensively used that they appear together many times.

We also noticed above that the tags which are most used are also the tags which are most viewed.The common tags are python,machine-learning,deep-learning, neural-network,keras,tensorflow,classification,scikit-learn.

The drawback of this method is that only consider pairing of one tag to another.But it fails to undertsnad relations between one group of tags to other group of tags.

For example it could be that scikit-learn and pandas together could have a strong relation with machine-learning while individually they don't.
To solve this we can consider some other techniques like [Association_Rules](https://en.wikipedia.org/wiki/Association_rule_learning).This technique will help us get the group of tags which appear together frequently.
