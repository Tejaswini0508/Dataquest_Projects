# Clean and Analyze Employee Exit Surveys

In this  project, we will work with exit surveys from employees of the [Department of Education](https://en.wikipedia.org/wiki/Department_of_Education_and_Training_(Queensland)), Training and Employment (DETE) and the Technical and Further Education (TAFE) institute in Queensland, Australia. You can find the TAFE exit survey [here](https://data.gov.au/dataset/ds-qld-89970a3b-182b-41ea-aea2-6f9f17b5907e/details?q=exit%20survey) and the survey for the DETE [here](https://data.gov.au/dataset/ds-qld-fe96ff30-d157-4a81-851d-215f2a0fe26d/details?q=exit%20survey).Some slight modifications have been made to these datasets to make them easier to work with, including changing the encoding to UTF-8 (the original ones are encoded using cp1252.)

In this project we will be analyzing the following:
* Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been there longer?
* Are younger employees resigning due to some kind of dissatisfaction? What about older employees?

We will be combining the results for both surveys to answer these questions. However, although both used the same survey template, one of them customized some of the answers.

From the above we can see that we need to convert both the columns so that they contain True or False only. Wherever there is text that column would be True or else it would be False.

We would not need to check the values in dete dataset since we have already seen above the data types for these columns is bool.Hence they will already be in the True/False form.

```python
#Updating the column values in tafe dataset
# Function to update values
def update_val(val):
    if pd.isnull(val)==True:
        return np.nan
    elif val=="-":
        return False
    else:
        return True


tafe_resignations_up=tafe_resignations.copy()
tafe_resignations_up["dissatisfied"]=tafe_resignations_up[["Contributing Factors. Dissatisfaction","Contributing Factors. Job Dissatisfaction"]].applymap(update_val).any(1,skipna=False)

tafe_resignations_up['dissatisfied'].value_counts(dropna=False)
```

```python
# Adding dissatisfied column in dete dataset

dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].any(1, skipna=False)
dete_resignations_up = dete_resignations.copy()

dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

As we can see above the column contains some absurd values. We will need to clean the data first to analyse. We will convert these numbers into categories.We'll base our analysis on  [this article](https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage), which makes the argument that understanding employee's needs according to career stage instead of age is more effective.

We will be uing the definitions below:

1. New: Less than 3 years at a company
2. Experienced: 3-6 years at a company
3. Established: 7-10 years at a company
4. Veteran: 11 or more years at a company

```python
# Extract the years of service from each value in institute_service column

combined["yearsofservice"]=combined["institute_service"].astype(str).str.extract(r"(\d+)").astype(float)
```

```python
# Map the yearsofservice value to one of the career stages above

def map_service(val):
    if val<3:
        return "Less than 3 years at a company"
    elif 3<=val<7:
        return "Experienced"
    elif 7<=val<=10:
        return "Established"
    elif pd.isnull(val)==True:
        return np.nan
    else:
        return "Veteran"
    

combined_updated["service_cat"]=combined_updated["yearsofservice"].apply(map_service)
```

```python
# Value counts in dissatisfied column
combined_updated["dissatisfied"].value_counts(dropna=False)

#Replace the Na's with most frequent value
combined_updated["dissatisfied"]=combined_updated["dissatisfied"].fillna(False)
```

```python
perc=combined_updated.pivot_table(values="dissatisfied",index="service_cat")
```

```python
import matplotlib as plt
%matplotlib inline

perc.plot(kind="bar")
```


From the above analysis we can conclude that employees who are Established i.e. they have compleyted 7 to 10 years in the company and employees who are Veteran i.e. they have completed less than 3 years are more likely to resign than the other employees because of some dissatisfaction they face at workplace.
